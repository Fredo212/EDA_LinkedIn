## **Introduction**

This Data Exploration Project will focus on the LinkedIn Job Postings Dataset. This dataset contains information on job postings on LinkedIn for the year 2023, covering job details, benefits, skills required, and more. From this dataset, I will be using four files:


*   The Job_postings CSV
*   The Job_skills CSV
*   The Company_industries CSV
*   The Skills CSV

The purpose of this analysis is to answer the following research questions:

*   Which industry offers the highest average salary?
*   What professional skills pay the most?

Through this analysis, I hope to gain insights into the current job market economy and how certain professional skills compare to others in financial value.

## **Data Preparation**

We will start the process by importing the necessary libraries and files that will be used in this project.

```python
import io
import pandas as pd
import plotly.express as px
df_postings = pd.read_csv('job_postings.csv')
df_skills = pd.read_csv('job_skills.csv')
df_industries = pd.read_csv('company_industries.csv')
df_skills_name = pd.read_csv("skills.csv")
```

```python
df_postings
```

```python
df_skills
```

```python
df_skills_name
```

```python
df_industries
```

Now, let's merge the files together

```python
df_1 = df_postings.merge(df_skills, how = "left", on="job_id")
```

```python
df_2= df_1.merge(df_industries, how = "left", on="company_id")
```

```python
df_final = df_2.merge(df_skills_name,how = "left", on="skill_abr")
```

```python
df_final
```

Now that we have merged our 4 files, it's time to clean our data so that we can make insights.

## **Data Cleaning**

Due to the nature of our merges, we ended up with a significant amount of duplicate values, so getting rid of them will be our first priority.

```python
df_clean = df_final.drop_duplicates()
df_clean
```

Notice that the amount of rows have significantly decreased from 463175 rows to 27604 rows.

Most of the columns in the dataset are irrelevant to the analysis we wish to conduct. Therefore, we have to remove them from the data frame.

For example, The skills CSV was used to have the full name of the abbreviations used in the job_skills CSV.

```python
df_clean = df_clean.drop(["description","applies","original_listed_time", "views", "job_posting_url","application_url","application_type","expiry",
               "closed_time", "formatted_experience_level", "listed_time" , "posting_domain", "sponsored",
               "work_type", "remote_allowed", "skills_desc", "compensation_type", "currency", "skill_abr"], axis = 1)
```

```python
df_clean
```

There seem to be fewer entries for median salaries compared to max/min salaries. We can calculate the median salaries by using the max/min salaries columns.

```python
df_clean["med_salary"] = (df_clean["max_salary"] + df_clean["min_salary"]) / 2
```

There seem to be salaries in three formats: yearly, monthly, and hourly. While we can convert the hourly and monthly salaries to yearly salaries, the yearly salaries make up the vast majority of the dataset, so we'll just focus on those only.

```python
df_clean = df_clean[df_clean["pay_period"] == "YEARLY"]
```

```python
df_clean
```

Now that all the entries in the dataset are showing yearly salaries, there's no need for the "pay_period" column.

```python
df_clean = df_clean.drop(["pay_period"], axis = 1)
```

```python
df_clean
```

Let's make the work_type column name shorter.

```python
df_clean = df_clean.rename(columns = {"formatted_work_type":"work_type", "skill_name": "skill"})
```

It's time to remove all the null values in our dataframe

```python
df_clean = df_clean.dropna()
```

Now let's check if the dataframe contains anymore null values.

```python
df_clean.isnull().sum()
```

Finally, the index of our dataframe does not reflect the number of rows it contains. This needs to be corrected.

```python
df_clean = df_clean.reset_index(drop = True)
```

Let's take a look at our now clean dataset.

```python
df_clean
```

Now that our dataset is cleaned, we can start our exploratory data analysis.

## **Exploratory Data Analysis (EDA)**

To answer our research question on which industries pay the most on average and which professional skills pay the most, we will use groupbys to analyze the industry and skills columns.

```python
industry = df_clean.groupby("industry")
industry_salary = industry.med_salary.mean().reset_index()
```

```python
fig = px.histogram(industry_salary, x= 'industry', y = "med_salary", title='Median Salary across Industries', marginal="box", hover_data = ['industry','med_salary'])
fig.show()
```

* It seems that the Online Media and Veterinary industries offer an extraordinary median salary compared to other industries. Let's take a look at these industries to make sure they are not outliers.

```python
df_clean[df_clean['industry'] == "Veterinary"]
```

Just as I thought, it looks like the Veterinary industry was an outlier.

```python
df_clean[df_clean['industry'] == "Online Media"]
```

Just like the Veterinary industry, it seems that the Online Media industry is also an outlier.

To make our data distribution more even, we'll have to remove those industries from the data.

```python
df_clean = df_clean.drop(df_clean[df_clean['industry'] == "Veterinary"].index)
df_clean = df_clean.drop(df_clean[df_clean['industry'] == "Online Media"].index)
```

Now let's check if the Veterinary industry is still present.

```python
df_clean[df_clean['industry'] == "Veterinary"]
```

Let's do the same for the Online Media industry.

```python
df_clean[df_clean['industry'] == "Online Media"]
```

Now that we removed those outliers, let's visualize the the median salaries by industry again.

```python
industry = df_clean.groupby("industry")
industry_salary = industry.med_salary.mean().reset_index()
```

```python
fig = px.histogram(industry_salary, x= 'industry', y = "med_salary", title='Median Salary across Industries', marginal="box", hover_data = ['industry','med_salary'])
fig.show()
```

 # **Insights:**

*   It seems the highest paying industry in this dataset when looking at their median salary is the Investment Banking industry with an average salary of about $205k.The next 4 top paying industries that follow are Outsourcing/Offshoring (193k), Public Policy (185k), Consumer Electronics(168k), and Semiconductors (160k).

*  I'm a little  surprised by the top 5 highest paying industries. I expected industries such as Consumer Electronics and Banking to by somewhere at the top. Public Policy, Outsourcing/Offshoring, and Semiconductors took me by surprise.

* Judging by the high figures I'm seeing for the median salaries, I'm guessing that the jobs associated with these skills are more senior positions rather than entry-level positions.

Now lets do the same thing for skill column.

```python
skills = df_clean.groupby("skill")
skills_salary = skills.med_salary.mean().reset_index()
```

```python
fig = px.histogram(skills_salary, x= 'skill', y = "med_salary", title='Median Salary across Professional Skills', marginal="box", hover_data = ['skill','med_salary'])
fig.show()
```
